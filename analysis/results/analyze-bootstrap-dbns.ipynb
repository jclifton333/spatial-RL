```python
import pickle as pkl
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import yaml
%matplotlib inline
```

```python
def display_results_from_yml(fname):
    d = yaml.load(open(fname))
    num_replicates = d['settings']['number_of_replicates']
    for i in range(num_replicates):
        d_rep = d['results'][i]
        for j in range(len(d['settings']['times_to_evaluate'])):
            sns.distplot(d_rep['mb_be'][j], label='mb')
            sns.distplot(d_rep['mf_be'][j], label='mf')
            print('Replicate: {} Time: {}'.format(i, d['settings']['times_to_evaluate'][j]))
            plt.legend()
            plt.show()
```

```python
fname1 = 'bootstrap_dbns_SIS_random_quad_approx_100_180722_113615.yml'
fname2 = 'bootstrap_dbns_SIS_random_quad_approx_100_180722_133250.yml'
fname3 = 'bootstrap_dbns_SIS_random_quad_approx_100_180722_133257.yml'
```

```python
display_results_from_yml(fname1)
```

```python
display_results_from_yml(fname2)
```

```python
def dropna(arr):
    return np.array(arr)[~np.isnan(arr)]
```

```python
omegas = [0, 0.5, 1]
replicates = range(5)
for i in range(len(omegas)):
    for j in range(len(replicates)):
        omega = omegas[i]
        replicate = replicates[j]
        fname = 'bootstrap-dbns-omega-{}-L-9-horizon-25-{}.p'.format(omega, replicate)
        df = pkl.load(open(fname, 'rb'))
        for t in range(len(df['mb_be'])):
            print('Omega: {} Replicate: {} Time: {}'.format(omega, replicate, df['time'][t]))
            sns.distplot(dropna(df['mb_be'][t]), label='mb')
            sns.distplot(dropna(df['mf_be'][t]), label='mf')
            plt.legend()
            plt.show()
```
