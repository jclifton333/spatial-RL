```python
def dropna(arr):
    return np.array(arr)[~np.isnan(arr)]
```

```python
omegas = [0, 0.5, 1]
replicates = [0, 2, 4]
for i in range(len(omegas)):
    for j in range(len(replicates)):
        omega = omegas[i]
        replicate = replicates[j]
        fname = 'bootstrap-dbns-omega-{}-L-9-horizon-50-{}.p'.format(omega, replicate)
        df = pkl.load(open(fname, 'rb'))
        for t in range(len(df['mb_be'])):
            print('Omega: {} Replicate: {} Time: {}'.format(omega, replicate, df['time'][t]))
            sns.distplot(dropna(df['mb_be'][t]), label='mb')
            sns.distplot(dropna(df['mf_be'][t]), label='mf')
            plt.legend()
            plt.show()
```
