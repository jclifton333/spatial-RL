```python
import pickle as pkl
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
omega00 = pkl.load(open('bootstrap-dbns-omega-0-horizon-50.p', 'rb'))
for t in range(len(omega00['mb_be'])):
    print('Time {}'.format(omega00['time'][t]))
    plt.hist(omega00['mb_be'][t], label='mb')
    plt.hist(omega00['mf_be'][t], label='mf')
    plt.legend()
    plt.show()
```

```python
omega05 = pkl.load(open('bootstrap-dbns-omega-0.5-horizon-50.p', 'rb'))
for t in range(len(omega05['mb_be'])):
    print('Time {}'.format(omega05['time'][t]))
    plt.hist(omega05['mb_be'][t], label='mb')
    plt.hist(omega05['mf_be'][t], label='mf')
    plt.legend()
    plt.show()
```

```python
omega10 = pkl.load(open('bootstrap-dbns-omega-1-horizon-50.p', 'rb'))
for t in range(len(omega10['mb_be'])):
    print('Time {}'.format(omega10['time'][t]))
    plt.hist(omega10['mb_be'][t], label='mb')
    plt.hist(omega10['mf_be'][t], label='mf')
    plt.legend()
    plt.show()
```

```python
colors = ['b', 'r', 'y']
dicts = [omega00, omega05, omega10]
for i in range(3):
    dict_ = dicts[i]
    plt.scatter(x=dict_['time'], y=[np.mean(dict_['mb_be'][t]) - np.mean(dict_['mf_be'][t])
                                   for t in range(len(dict_['mb_be']))],
                c=colors[i])
```
