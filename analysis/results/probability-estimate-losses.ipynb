```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import yaml
import pandas as pd
%matplotlib inline
```

```python
def plot_losses_over_time(fname):
    d = yaml.load(open(fname))
    names = d['results'][0].keys()
    
    # Collect series
    loss_names = []
    losses = []
    times = []
    reps = []
    for rep, rep_results in d['results'].items():
        for k, v in rep_results.items():
            losses += v
            loss_names += [k]*len(v)
            times += [i for i in range(len(v))]
            reps += [rep]*len(v)
    df = pd.DataFrame.from_dict({'loss_names': loss_names, 'losses': losses, 'times': times,
                                 'reps': reps})
    df.set_index('times', inplace=True)
    df.groupby('loss_names')['losses'].plot(legend=True)
    return
```

```python
plot_losses_over_time('prob_estimates_SIS_random_quad_approx_50_0.0_180728_003732.yml')
```
