```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import yaml
%matplotlib inline
```

```python
policy_names_dict = {'one_step': 'One-step MF', 'random': 'Random', 'SIS_model_based_one_step': 'One-step MB',
                    'sis_one_step_mse_averaged': 'One-step MSE-averaged'}

def plot_results_in_list(file_list, horizon=None, L=None, title=None):
    for f in file_list:
        d = yaml.load(open(f))
        results = [r['score'] for r in d['results'].values() if isinstance(r, dict)]
        policy_name = d['settings']['policy_name']
        if policy_name in policy_names_dict.keys():
            policy_name = policy_names_dict[policy_name]
        sns.distplot(results, label=policy_name)
    plt.legend(bbox_to_anchor=(0, 1.05))
    plt.xlabel('Average number of infections')
    if title is not None:
        if horizon is not None and L is not None:
            title += '\nSteps per episode={}\nNumber of locations={}'.format(horizon, L)
        plt.title(title)
    plt.show()
```

```python
file_list0 = ['SIS_rollout_quad_approx_50_0.0_180727_085730.yml', 
             'SIS_random_quad_approx_50_0.0_180727_084927.yml',
             'SIS_SIS_model_based_one_step_quad_approx_50_0.0_180727_214237.yml']
```

```python
file_list1 = ['SIS_random_quad_approx_50_1.0_180727_084927.yml',
              'SIS_rollout_quad_approx_50_1.0_180727_085730.yml',
              'SIS_SIS_model_based_one_step_quad_approx_50_1.0_180727_214555.yml']
```
