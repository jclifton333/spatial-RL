```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import yaml
%matplotlib inline
```

```python
def plot_results_in_list(file_list):
    for f in file_list:
        d = yaml.load(open(f))
        results = [r['score'] for r in d['results'].values() if isinstance(r, dict)]
        sns.distplot(results, label='{}-{}'.format(d['settings']['policy_name'],
                                                  d['settings']['rollout_depth']))
    plt.legend(bbox_to_anchor=(0, 1.05))
    plt.show()
```

```python
file_list0 = ['SIS_rollout_quad_approx_50_0.0_180727_085730.yml', 
             'SIS_random_quad_approx_50_0.0_180727_084927.yml',
             'SIS_SIS_model_based_quad_approx_50_0.0_180727_172016.yml',
             'SIS_SIS_model_based_one_step_quad_approx_50_0.0_180727_214237.yml']
```
